Task 2:Build Advanced Preprocessing Function

```python
import re
from collections import Counter

def preprocess_text(text):
    if not text or not isinstance(text, str):
        return [], ""

    # 1. Lowercase
    text = text.lower()

    # 2. Remove URLs & emails
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)

    # 3. Remove numbers
    text = re.sub(r'\d+', '', text)

    # 4. Remove repeated characters (soooo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # 5. Remove special characters & emojis
    text = re.sub(r'[^a-z\s]', '', text)

    # 6. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # 7. Tokenization
    tokens = text.split()

    # 8. Remove very short tokens (≤2) except 'no', 'not'
    cleaned_tokens = [
        word for word in tokens
        if len(word) > 2 or word in ["no", "not"]
    ]

    # Final cleaned sentence
    cleaned_sentence = " ".join(cleaned_tokens)

    return cleaned_tokens, cleaned_sentence
```

Task 3: Stress Testing

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for sentence in test_sentences:
    # Pass the entire text to preprocess function
    tokens, clean_sentence = preprocess_text(sentence)

    print("\n==============================")
    print()
    print("Original:", sentence)
    print("Tokens:", tokens)
    print("Clean Sentence:", clean_sentence)

    results.append((sentence, tokens, clean_sentence))
```

Task 4: Token Analytics

```python
print("\nTOKEN ANALYTICS")

for original, tokens, clean_sentence in results:
    # Length of tokens
    total_tokens = len(tokens)

    # Total Unique tokens
    unique_tokens = len(set(tokens))

    # calcukate the avg
    avg_length = sum(len(word) for word in tokens) / total_tokens if total_tokens > 0 else 0

    print("\n------------------------------")
    print()
    print("Sentence:", original)
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Average Token Length:", round(avg_length, 2))
```

Task 5: Frequency Analysis

```python
#Create the empty list to store every word
all_tokens = []

for _, tokens, _ in results:
    all_tokens.extend(tokens)

# Count how many time unique word appearance
counter = Counter(all_tokens)

# Display Top 10
print("\nTop 10 Most Frequent Words:")
print('-'*150)
print(counter.most_common(10))

# Display Top 5
print("\nTop 5 Least Frequent Words:")
print('-'*150)
least_common = sorted(counter.items(), key=lambda x: x[1])
print(least_common[:5])
```

Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        # Preproceess the text
        tokens, sentence = preprocess_text(text)

        # Add new word to it
        all_tokens.extend(tokens)

        # Cleaned-up sentence
        clean_sentences.append(sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
# Pass the text to pipeline
output = full_pipeline(test_sentences)

print("\nFULL PIPELINE OUTPUT")
print('-'*150)
print()

# Display the tokens
print("Tokens:", output["tokens"])

# Display the cleaned sentences
print("\nClean Sentences:", output["clean_sentences"])
```

Task 7: Error Handling

```python
# List of empty string ,emoji's and numbers
edge_cases = [
    "",
    "😍😍😍😂😂",
    "1234567890"
]

print("\nERROR HANDLING TEST")

# loop each test case
for case in edge_cases:

    # Preprocess the edge cases
    tokens, sentence = preprocess_text(case)

    print("\n----------------------")
    print()
    print("Input:", case)
    print("Tokens:", tokens)
    print("Clean Sentence:", sentence)
```
